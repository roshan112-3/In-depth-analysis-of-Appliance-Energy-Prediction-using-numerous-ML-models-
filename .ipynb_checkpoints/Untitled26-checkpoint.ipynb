```python
import numpy as np
import pandas as pd 
import seaborn as sns
import datetime as dt
from datetime import datetime
from matplotlib import pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import random
```

```python
energy = pd.read_csv(r"KAG_energydata_complete.csv")
energy.head()
```

```python
#converting date into datetime
energy['date'] = energy['date'].astype('datetime64[ns]')
energy.head()
```

```python
#making the date, time, hour and month into a diff column and thus converting them into date time format
energy['Date'] = pd.to_datetime(energy['date']).dt.date
energy['Time'] = pd.to_datetime(energy['date']).dt.time
energy['hour'] = energy['date'].dt.hour
energy['month'] = energy['date'].dt.month
energy['Day'] = pd.to_datetime(energy['date']).dt.date
```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame(energy)
df['Date'] = pd.to_datetime(df['Date'])

# group the data by month and create a box plot for each group
df['Month'] = df['Date'].dt.month_name()
sns.set(rc={'figure.figsize':(12,6)}) # set the figure size
sns.boxplot(x='Month', y='Appliances', data=df)
```

```python
df = pd.DataFrame(energy)
df['Date'] = pd.to_datetime(df['Date'])

# create subplots for each month
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(6, 12), sharex=True)
months = df['Date'].dt.month_name().unique()
for i, month in enumerate(months):
    data_month = df[df['Date'].dt.month_name() == month]
    axs[i].plot(data_month['Date'], data_month['Appliances'])
    axs[i].set_title(month)
    axs[i].set_ylabel('Appliances')
plt.xlabel('Date')
plt.show()



```

```python
df = pd.DataFrame(energy)
df['date'] = pd.to_datetime(df['date'])

# create grid of subplots
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 4))

# loop through subplots and plot data for corresponding month
months = df['date'].dt.month.unique()
for i, month in enumerate(months):
    data_month = df[df['date'].dt.month == month]
    axs[i].plot(data_month['date'], data_month['Appliances'])
    axs[i].set_title(f'Month {month}')
    axs[i].set_xlabel('')
    axs[i].set_ylabel('Appliances')

# add overall title and show plot
fig.suptitle('Value by Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# assume energy is a list of dictionaries with 'date' and 'Appliances' keys

df = pd.DataFrame(energy)
df['date'] = pd.to_datetime(df['date'])

# create figure and axes
fig, ax = plt.subplots(figsize=(13, 7))

# plot data by month
for month in df['date'].dt.month.unique():
    data_month = df[df['date'].dt.month == month]
    ax.plot(data_month['date'], data_month['Appliances'], marker='o', linestyle='--', label=f'Month {month}')

# format x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.set_xlim(df['date'].min(), df['date'].max())

# add legend and title
ax.legend()
ax.set_title('Monthly Data')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

df = pd.DataFrame(energy)

# plot data for January
jan_data = df[df['date'].dt.month == 1]
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(jan_data['date'], jan_data['Appliances'], marker='o', linestyle='--', label='January')

# set x-axis major tick locator to display days
ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))

# set x-axis minor tick locator to display hours
ax.xaxis.set_minor_locator(mdates.HourLocator(interval=6))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%H:%M'))

# add labels and legend
ax.set_xlabel('Date')
ax.set_ylabel('Appliances')
ax.set_title('Monthly Appliances Usage - January')
ax.legend()

# rotate x-axis labels
plt.xticks(rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(energy)

# group data by month
monthly_data = df.groupby(df['date'].dt.strftime('%B'))['Appliances'].sum()

# plot data
fig, ax = plt.subplots(figsize=(12, 6))
monthly_data.plot(kind='bar', ax=ax, color=['blue','orange','green','red','violet'])

# set labels and title
ax.set_xlabel('Month')
ax.set_ylabel('Total Appliances Usage')
ax.set_title('Monthly Appliances Usage')

plt.show()
```

```python
#divide the columns based on type
col_temp = ["T1","T2","T3","T4","T5","T6","T7","T8","T9"]
col_hum = ["RH_1","RH_2","RH_3","RH_4","RH_5","RH_6","RH_7","RH_8","RH_9"]
col_weather = ["T_out", "Tdewpoint","RH_out","Press_mm_hg","Windspeed","Visibility"] 
col_light = ["lights"]
col_randoms = ["rv1", "rv2"]
col_target = ["Appliances"]
```

```python
#scaling the data
scaler = MinMaxScaler(feature_range=(0,10))
energy[['lights','T1','RH_1','T2','RH_2','T3','RH_3','T4','RH_4','T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8','T9','RH_9','T_out','Press_mm_hg','RH_out','Windspeed','Visibility','Tdewpoint','rv1','rv2']] = scaler.fit_transform(energy[['lights','T1','RH_1','T2','RH_2','T3','RH_3','T4','RH_4','T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8','T9','RH_9','T_out','Press_mm_hg','RH_out','Windspeed','Visibility','Tdewpoint','rv1','rv2']])
energy_transformed = energy
energy_transformed
#splitting into test and train
train, test = train_test_split(energy_transformed,test_size=0.30,random_state=25)
# Seperate dependent and independent variables 
feature_vars = train[col_temp + col_hum + col_weather + col_light + col_randoms]
target_vars = train[col_target]
```
