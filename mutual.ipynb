```python
energy = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\KAG_energydata_complete.csv")
energy.head()
```

```python
#converting date into datetime
energy['date'] = energy['date'].astype('datetime64[ns]')
energy.head()
```

```python
#making the date, time, hour and month into a diff column and thus converting them into date time format
energy['Date'] = pd.to_datetime(energy['date']).dt.date
energy['Time'] = pd.to_datetime(energy['date']).dt.time
energy['hour'] = energy['date'].dt.hour
energy['month'] = energy['date'].dt.month
```

```python
energy.drop(['Date', 'Time'], axis = 1 , inplace = True)
```

```python
energy1 = energy.iloc[ : , 1:31]
energy1
```

```python
#creating feature vector
feature = energy1.iloc[:, 1:30]
feature
```

```python
#splitting into test and train
X = feature
y = energy['Appliances']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=40)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, RepeatedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_selection import mutual_info_regression, SelectKBest
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import Lars, Ridge, LinearRegression, BayesianRidge, Lasso, ElasticNet, PassiveAggressiveRegressor, HuberRegressor, LassoLars
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
import numpy as np



# Define the regressors to evaluate
regressors = [ExtraTreesRegressor(),
              RandomForestRegressor(),
              XGBRegressor(),
              LGBMRegressor(),
             ]

# Define the parameter grid for the number of features to select
param_grid = {'select__k': range(1, X.shape[1] + 1)}

# Define the cross-validation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# Define the evaluation metrics
metrics = ['neg_mean_squared_error', 'neg_mean_absolute_error', 'r2', 'neg_mean_absolute_percentage_error']

feature_names = ('lights','T1','RH_1','T2','RH_2','T3','RH_3','T4','RH_4',
                 'T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8','T9','RH_9','T_out',
                 'Press_mm_hg','RH_out','Windspeed','Visibility','Tdewpoint','rv1','rv2', 'hour', 'month')

# Loop over the regressors and evaluate them using feature selection and hyperparameter tuning
for regressor in regressors:
    # Define the feature selection and multiple regression pipeline
    pipe = Pipeline([('select', SelectKBest(score_func=mutual_info_regression)),
                     ('regress', regressor)])
    
    # Define the GridSearchCV object
    grid = GridSearchCV(pipe, param_grid=param_grid, cv=cv, scoring=metrics, refit='neg_mean_squared_error', return_train_score=True)
    
    # Fit the grid search to the data
    grid.fit(X_train, y_train)
    
    # Get the best number of features
    best_k = grid.best_params_['select__k']
    
    # Get the selected feature indices and names
    selected_indices = grid.best_estimator_.named_steps['select'].get_support(indices=True)
    selected_features = [feature_names[i] for i in selected_indices]
    
    # Print the best number of features and the selected feature names
    print("Regressor:", type(regressor).__name__)
    print("Best number of features:", best_k)
    print("Selected features:", selected_features)
    
    # Get the best estimator
    best_estimator = grid.best_estimator_
    
    # Get the predictions on the train and test sets
    y_train_pred = best_estimator.predict(X_train)
    y_test_pred = best_estimator.predict(X_test)

    # Calculate the evaluation metrics for train set
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mape = np.mean(np.abs((y_train - y_train_pred) / y_train)) * 100
    train_r2 = r2_score(y_train, y_train_pred)

    # Calculate the evaluation metrics for test set
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mape = np.mean(np.abs((y_test - y_test_pred) / y_test)) * 100
    test_r2 = r2_score(y_test, y_test_pred)

    # Print evaluation metrics for both train and test sets
    print("Train set - RMSE:", train_rmse, "MAE:", train_mae, "MAPE:", train_mape, "R2:", train_r2)
    print("Test set - RMSE:", test_rmse, "MAE:", test_mae, "MAPE:", test_mape, "R2:", test_r2)
    print("\n")
```

```python
import matplotlib.pyplot as plt

# Plot feature importances for tree-based models
if isinstance(regressor, DecisionTreeRegressor) or isinstance(regressor, ExtraTreesRegressor) or isinstance(regressor, RandomForestRegressor) or isinstance(regressor, GradientBoostingRegressor) or isinstance(regressor, AdaBoostRegressor) or isinstance(regressor, LGBMRegressor) or isinstance(regressor, XGBRegressor):
    importances = best_estimator.named_steps['regress'].feature_importances_
    indices = np.argsort(importances)[::-1]
    plt.figure()
    plt.title(type(regressor).__name__ + ' Feature Importances')
    plt.barh(range(X.shape[1]), importances[indices], align='center')
    plt.yticks(range(X.shape[1]), [feature_names[i] for i in indices])
    plt.xlabel('Relative Importance')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Define a list of metric names
metric_names = ['RMSE', 'MAE', 'MAPE', 'R2']

# Create subplots for each metric
fig, axes = plt.subplots(nrows=1, ncols=len(metric_names), figsize=(20, 5))

metrics = ['neg_mean_squared_error', 'neg_mean_absolute_error', 'r2', 'neg_mean_absolute_percentage_error', 'neg_root_mean_squared_error']


# Loop over the metrics and create bar plots
for i, metric in enumerate(metric_names):
    # Get the metric values for each regressor
    train_metric_values = [np.sqrt(-score) if metric == 'RMSE' else -score for score in grid.cv_results_['mean_train_' + metric]]
    test_metric_values = [np.sqrt(-score) if metric == 'RMSE' else -score for score in grid.cv_results_['mean_test_' + metric]]
    
    # Plot the bar chart
    ax = axes[i]
    ax.bar(np.arange(len(regressors)), train_metric_values, label='Train', color='blue')
    ax.bar(np.arange(len(regressors))+0.4, test_metric_values, label='Test', color='orange')
    ax.set_xticks(np.arange(len(regressors))+0.2)
    ax.set_xticklabels([type(regressor).__name__ for regressor in regressors])
    ax.set_xlabel('Regressor')
    ax.set_ylabel(metric)
    ax.legend()

plt.show()
```
